```python
import re
import requests
from urllib.request import urlopen
from bs4 import BeautifulSoup
import urllib

import pandas as pd
```

```python
team = pd.read_csv("2020빅콘테스트_스포츠투아이_제공데이터_팀_2020.csv", engine = 'python')
```

```python
tmp1 = list(team.T_NM)
tmp2 = list(team.T_ID)

team_dict = dict()

for i in range(len(tmp1)):
    team_dict[tmp1[i]] = tmp2[i]
team_dict
```

### 1. 함수

```python
def read_txt(file_name):

    # 날짜 데이터 읽기
    f = open(file_name, mode = 'r')
    f_urls = f.readlines()
    urls = f_urls[0][1:-1].split(',')
    return urls
```

```python
def make_dataframe(file_name):
    
    col_lst = ['GDAY_DS','T_ID']
    
    if file_name == '득점권_투수2.txt':
        col_lst.append('WHIP')
    elif file_name == '득점권_타자2.txt':
        col_lst.append('득점권_타율')

    # dataframe
    data = pd.DataFrame([],columns = col_lst)

    return data
```

```python
def create_row(data, urls, num):
    idx = 0
    for url in urls:
        url = url.strip()[1:-1]

        # 날짜
        day_lst = url.split('&')[28].split('=')[1].split('-')
        day = '2020{}{}'.format(day_lst[0],day_lst[1])
        print("=====",day)

        fp = urlopen(url)
        source = fp.read()
        fp.close()

        soup = BeautifulSoup(source)

        table = soup.find(id='mytable')
        trs = table.find_all('tr')  

        for i in range(3,len(trs)):
            tmp = trs[i].text.split()

            for t in team_dict:
                if t in tmp[0]:
                    t_id = team_dict[t]
                    break

            row = [day, t_id]
            
            value = tmp[num]

            row.append(value)
            
            print("ROW",value, row)

            data.loc[idx,:] = row
            idx += 1

    return data
```

### 2. 함수사용

#### 1) 득점권 투수 

```python
file_name = '득점권_투수2.txt' # file 명 따라 수정
urls = read_txt(file_name)
pp_data = make_dataframe(file_name)
sco_p2020 = create_row(pp_data, urls, -5)
```

```python
sco_p2020
```

```python
sco_p2020_fill = sco_p2020.fillna(0)
```

```python
ppp = pd.read_csv("pitcher2020_2.csv")
ppp['GDAY_DS'] = ppp['GDAY_DS'].astype(str)
tt = pd.merge(ppp, sco_p2020, how = 'left', left_on = ['GDAY_DS','T_ID'], right_on = ['GDAY_DS','T_ID'])
tt[tt["WHIP"].isnull()]
```

```python
save_pp = pd.merge(ppp, sco_p2020_fill, how = 'left', left_on = ['GDAY_DS','T_ID'], right_on = ['GDAY_DS','T_ID']).sort_values(by = ["GDAY_DS","T_ID"]).reset_index(drop=True)
save_pp
```

```python
save_pp.to_csv("0920_total_pitcher_2.csv",index=False)
```

```python
file_name = '득점권_타자2.txt' # file 명 따라 수정
urls = read_txt(file_name)
hh_data = make_dataframe(file_name)
sco_h2020 = create_row(hh_data, urls, -4)
```
