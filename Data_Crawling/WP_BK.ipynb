```python
import pandas as pd
```

### 1. BK

** Excel 자체에서 GDAY 채워주었습니다

```python
bk = pd.read_csv("BK.csv")
bk["GDAY_DS"] = bk["GDAY_DS"].astype('object')
bk.head()
```

```python
bk_new = bk.melt(id_vars = ['GDAY_DS',"HEADER_NO"], var_name = "T_ID", 
        value_name = "BK").dropna(axis = 0)

bk_new = bk_new.sort_values(by = ["GDAY_DS","T_ID"]).reset_index(drop = True)
bk_new["BK"]= bk_new["BK"].astype('int64')
```

```python
bk_new.to_csv("BK_new.csv",index = False)
```

### 2. WP

```python
wp = pd.read_csv("WP.csv")
wp["GDAY_DS"] = wp["GDAY_DS"].astype('object')
wp.head()
```

```python
wp_new = wp.melt(id_vars = ['GDAY_DS',"HEADER_NO"], var_name = "T_ID", 
        value_name = "WP").dropna(axis = 0)

wp_new = wp_new.sort_values(by = ["GDAY_DS","T_ID"]).reset_index(drop = True)
wp_new["WP"]= wp_new["WP"].astype('int64')
```
